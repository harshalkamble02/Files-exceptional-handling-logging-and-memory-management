Q1. What is the difference between interpreted and compiled languages?
-  - Compiled: Translated into machine code first (e.g., C, C++). Runs faster.

 - Interpreted: Runs line by line by an interpreter (e.g., Python). Easier to debug but slower.

Q2.What is exception handling in Python?
-  A mechanism to handle runtime errors gracefully using try, except, finally, and else.

Q3.What is the purpose of the finally block in exception handling?
- Code inside finally always executes, even if an error occurs (used for cleanup like closing files).

Q4.What is logging in Python?
- A way to record events, errors, and messages during program execution using the logging module.

Q5.What is the significance of the __del__ method in Python?
- A destructor method that is called when an object is deleted or garbage collected.

Q6.What is the difference between import and from ... import in Python?
-  - import module → brings the whole module.

 - from module import function → brings specific parts only.

Q7.How can you handle multiple exceptions in Python?
- Use multiple except blocks or a single except (Error1, Error2) block.

Q8.What is the purpose of the with statement when handling files in Python?
- It automatically closes the file after operations, even if an error occurs.


Q9.What is the difference between multithreading and multiprocessing?
- -  Multithreading: Runs multiple threads in one process (shares memory).

 - Multiprocessing: Runs multiple processes (separate memory spaces).

Q10.What are the advantages of using logging in a program?
- Helps in debugging, keeps a record of events, and allows monitoring without stopping the program.

Q11.What is memory management in Python?
- Process of allocating, using, and freeing memory efficiently (handled by Python’s garbage collector).

Q12.What are the basic steps involved in exception handling in Python?
- - Wrap risky code in try.

 - Catch errors in except.

 - Use else (optional) for no-error code.

 - Use finally for cleanup.


Q13.Why is memory management important in Python?
- Prevents memory leaks, improves performance, and ensures efficient resource usage.

Q14.What is the role of try and except in exception handling?
-  - try → tests code for errors.

 - except → catches and handles the error.



Q15.How does Python's garbage collection system work?
- Uses reference counting and generational garbage collection to free unused memory automatically.

Q16.What is the purpose of the else block in exception handling?
- Runs only if no exception occurs in the try block.


Q17.What are the common logging levels in Python?
- DEBUG, INFO, WARNING, ERROR, CRITICAL (from lowest to highest severity).

Q18.What is the difference between os.fork() and multiprocessing in Python?
-  - os.fork() → Unix-only, low-level process creation.

  - multiprocessing → Cross-platform, higher-level, easier to manage.


Q19.What is the importance of closing a file in Python?
- Frees system resources and ensures all data is saved properly.


Q20.What is the difference between file.read() and file.readline() in Python?
- - read() → Reads the whole file.

 - readline() → Reads one line at a time.

Q21.What is the logging module in Python used for?
 - To log messages (info, warning, errors) for debugging and tracking.

Q22.What is the os module in Python used for in file handling?
- Provides OS-level functions like file paths, checking existence, deleting, and renaming files.

Q23.What are the challenges associated with memory management in Python.
- Reference cycles, memory fragmentation, and managing large objects can be tricky.

raise ValueError("Invalid input")
